```python
class Employee:
    def __init__(self,name,position,salary):
        self.name=name
        self.position=position
        self.salary=salary
    
    def display_info(self):
        return f"{self.name} is a {self.position} earning ${self.salary} annually"

employee1=Employee('Amit','Senior Software Developer',30000)
print(employee1.display_info())
employee2=Employee('Bandana Rani','Data Engineer',34567)
print(employee2.display_info())
employee2.salary=40000
print(employee2.display_info())
```
