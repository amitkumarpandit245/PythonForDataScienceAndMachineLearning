```python
fruits=['apple','mango','Banana']
for fruit in fruits:
    print(fruit)
```

```python
it=iter(fruits)
print(next(it))
print(next(it))
print(next(it))
```

```python
def is_iterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:
        return False
```

```python
print(is_iterable(fruits))
```

```python
print(is_iterable(10))
```

```python
numbers=[1,2,3,4,5,6,7,8]
even_num=filter(lambda x:x%2 ==0, numbers)
print(list(even_num))
```

```python
squared_list=map(lambda x : x**2,numbers)
print(list(squared_list))
```

```python
odd_squared=map(lambda x: x**2,list(filter(lambda x: x%2 != 0,numbers)))
print(list(odd_squared))
```

```python
numbers= [1,-1,0,-2,3,4,5,6]
print(all(num>0 for num in numbers))
```
