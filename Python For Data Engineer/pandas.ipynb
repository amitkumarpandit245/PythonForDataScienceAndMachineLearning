```python
import pandas as pd
```

```python
data_excel=pd.read_excel('Loan_Data.xlsx')
print(data_excel.to_string())
```

```python
#data cleaning and preprocess
data_branch_per=pd.read_csv('Branch_Performance.csv')
print(data_branch_per.to_string())
```

```python
new_data_branch_per=data_branch_per.dropna()
print(new_data_branch_per.to_string())
```

```python

df=data_branch_per.fillna(100)
print(df.to_string())
```

```python
data_branch_per['Region'].fillna('AMIT',inplace=True)
print(data_branch_per.to_string())
```

```python
df_new=data_branch_per.drop_duplicates()
print(df_new.to_string())
```

```python
data_excel['TenureMonths']=pd.to_datetime(data_excel['TenureMonths'],format='%d%m%Y',errors='ignore')
print(data_excel.to_string())
```

```python
avg_Loan=data_excel.groupby('LoanType')['LoanAmount'].mean()
print(avg_Loan)
```

```python
data_excel.aggregate(['sum','min'])
```

```python
data_excel['LoanAmount'].aggregate(['sum','min'])
```

```python
#Merging File
new_data_excel=pd.read_excel('Loan_Data.xlsx')
merged_data=pd.merge(data_excel,new_data_excel,on='LoanID')
print(merged_data)
```

```python
merged_data.to_csv('Output.csv')
```

```python
merged_data.to_excel('OutputExcel.xlsx')
```

```python
merged_data.to_json('OutputJson.json')
```
