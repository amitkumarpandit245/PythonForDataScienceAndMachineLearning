```python
import requests
from bs4 import BeautifulSoup
url='https://azurelib.com/azure-interview-questions-and-answers/'
response=requests.get(url)
```

```python
html_content=response.content
print(html_content)
```

```python
soup=BeautifulSoup(html_content,'html.parser')
h1_tags=soup.find('h2')
print(h1_tags.text)
```

```python
imgs=soup.find_all('img')
for img in imgs:
    print(img['src'])
```

```python
myList=soup.find('ul')
print(myList.text)
```

```python
text= myList.text
with open('output_web_scrap.txt','w') as file:
    file.write(text)
```
