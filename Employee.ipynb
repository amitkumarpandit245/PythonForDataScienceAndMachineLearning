```python
import pyodbc

conn_str = (
    "DRIVER={PostgreSQL ODBC Driver(UNICODE)};"
    "SERVER=localhost;"
    "PORT=5432;"
    "DATABASE=amba;"
    "UID=postgres;"
    "PWD=amit@123;"
)

conn = pyodbc.connect(conn_str)
cursor = conn.cursor()

print("Connected Successfully!")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS employee1 (
    id SERIAL PRIMARY KEY,
    name VARCHAR(100),
    salary NUMERIC
);
""")
conn.commit()
print("Table created")
```

```python
cursor.execute("""
insert into employee values('Bandana Rani',34,'Hajipur','Engineer')
""")
conn.commit()
```

```python
cursor.execute("""
select * from employee1;
""")
rows = cursor.fetchall()
print(rows)
```
