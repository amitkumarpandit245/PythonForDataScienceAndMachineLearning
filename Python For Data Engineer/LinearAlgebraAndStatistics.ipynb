```python
import numpy as np
array1=np.array(([1,2],[3,4]))
array2=np.array(([5,6],[7,8]))
mult=np.dot(array1,array2)
print(mult)
```

```python
inv=np.linalg.inv(array1)
print(inv)
```

```python
eval, evect=np.linalg.eig(array1)
print(eval)
print(evect)
```

```python
data=np.array([1,2,3,4,5,6,7])
print(np.mean(data))
print(np.std(array1))
```

```python
data2=np.array([11,12,13,14,15,16,17])
cor=np.corrcoef(data,data2)
print(cor)
```
