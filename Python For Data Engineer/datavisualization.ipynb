```python
import matplotlib.pyplot as plt
```

```python
data= {'Date':['2024-08-01','2023-08-01','2022-08-01'],'Price':[500,200,300]}
df=pd.DataFrame(data)
df['Date']=pd.to_datetime(df['Date'])
df.plot(x='Date',y='Price',kind='line')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
data={'Product':['Apple','Banana','Grapes'],"Sales":[23,45,67]}
df=pd.DataFrame(data)
df.plot(x='Product',y='Sales',kind='bar')
plt.show()
```

```python
data={'Hours_Studied':[2,3,4,5,6,7],'Exam_Score':[200,300,400,500,200,700]}
df=pd.DataFrame(data)
df.plot(x='Hours_Studied',y='Exam_Score',kind='scatter')
plt.show()
```

```python
data={'Toppings':['Pepperoni','Mushroom','Olive','Sausage'],'Count':[23,4,56,56]}
df=pd.DataFrame(data)
df.plot(x='Toppings',y='Count',kind='pie')
plt.show()
```
